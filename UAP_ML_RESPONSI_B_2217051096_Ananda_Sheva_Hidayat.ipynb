# Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

# Load Dataframe

```python
#load data
df = pd.read_csv('https://raw.githubusercontent.com/alfarzf/datasets/main/data.csv')
df
```

# EDA

```python
#info dataframe
df.info()
```

```python
#menghitung nilai NaN
df.isnull().sum()
```

```python
df.describe()
```

# Preprocessing

```python
# drop kolom NaN
X = df.drop(['workclass', 'occupation','native-country'], axis=1)
y = df['income']
```

```python
sc = StandardScaler()
X = X.select_dtypes(include=['number'])
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

# Modelling

```python
classifier = SVC(kernel='linear', random_state=2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print("Accuracy_score :", accuracy_score(y_test, y_pred))
```

```python
scores = cross_val_score(classifier, X, y, cv=10)
print("Accuracy: ",round(scores.mean(), 2) * 100, "%")
print("Standard Deviation: +/-", scores.std())
```

# Evaluation

```python
scores2 = cross_val_score(classifier, X_train, y_train, cv = 10)
```

```python
print("Accuracy: ", round(scores2.mean(), 2) * 100, "%")
print("Standard Deviation: +/-", scores2.std())
```
